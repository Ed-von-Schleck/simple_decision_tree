```python
def entropy(X):
    return -sum(p * math.log(p) for p in X)
```

```python
def calculate_impurity(targets):
    """Calculate the impurity from a list of targets
    
    Args:
        targets (list): A list of target labels (e.g. ["a", "a", "b", "a", "c", ])
        
    Returns:
        (float): The impurity (messiness) of the label distribution
    """
    
    X = [count / len(targets) for count in collections.Counter(targets).values()]
    return entropy(X) * len(targets)
```
