```python
def entropy(X):
    return -sum(p * math.log(p) for p in X)
```

```python
def make_target_distribution(targets):
    counts = collections.Counter(targets)
    total = len(targets)
    return [count / total for count in counts.values()]
```

```python
def calculate_impurity(targets):
    """Calculate the impurity from a list of targets
    
    Args:
        targets (list): A list of target labels
        
    Returns:
        (float): A
    """
    X = make_target_distribution(targets)
    return entropy(X) * len(targets)
```

```python
def split(features, targets, attr):
    feature_partitions = [[] for _ in range(len(ATTRS[attr]))]
    target_partitions = [[] for _ in range(len(ATTRS[attr]))]
    
    for feats, target in zip(features, targets):
        index = ATTRS[attr].index(feats[attr])
        feature_partitions[index].append(feats)
        target_partitions[index].append(target)
        
    return feature_partitions, target_partitions
```

```python
def find_best_split(features, targets):
    best_impurity = None
    best_attr = None
    best_target_partitions = None
    best_feature_partitions = None
    
    for attr in range(len(ATTRS)):
        feature_partititions, target_partitions = split(features, targets, attr)
        impurity = sum(calculate_impurity(targets) for targets in target_partitions)
        
        if best_impurity is None or impurity < best_impurity:
            best_impurity = impurity
            best_attr = attr
            best_target_partitions = target_partitions
            best_feature_partitions = feature_partititions
            
    return best_attr, best_impurity, best_feature_partitions, best_target_partitions
```

```python
def build_tree(features, targets):
    current_impurity = calculate_impurity(targets)
    best_attr, best_impurity, best_feature_partitions, best_target_partitions = find_best_split(features, targets)

    if best_impurity < current_impurity:
        return (
            best_attr,
            [build_tree(f, t) for f, t in zip(best_feature_partitions, best_target_partitions)]
        )
    else:
        return collections.Counter(targets)
```

```python
decision_tree = build_tree(features, targets)
```
